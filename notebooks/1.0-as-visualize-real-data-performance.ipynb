```python
import os
import sys
# path to settings
sys.path.insert(0, os.path.split(os.getcwd())[0])
from settings import REAL_DATA_PREDICTIONS, CUT_DIR
from random import randint
import h5py
from sty import fg, bg, rs


def get_colors():
    class_labels = os.listdir(CUT_DIR)
    class_labels.sort()

    COLORS = {}
    for i, label in enumerate(class_labels):
        COLORS[label] = '%s%s %s %s%s' % (bg(randint(0, 254)), fg.black, label, fg.rs, bg.rs)

    return COLORS


def format_output(predictions_list, colors_dict):
    formated_string = ''

    for prediction in predictions_list:
        formated_string += colors_dict[prediction]

    return formated_string


colors_dict = get_colors()
for filename in os.listdir(REAL_DATA_PREDICTIONS):
    print("Filename: %s" % (filename))
    file = h5py.File(os.path.join(REAL_DATA_PREDICTIONS, filename), 'r')
    print(format_output(file['predictions'].value, colors_dict))
    print("\n\n")
```
