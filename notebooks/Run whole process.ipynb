```python
def create_data(timeframe):
    # 1. step cut recordings
    print('1. Cutting recordings')
    subprocess.call(
        ['python', 'cut_original_recordings.py', str(timeframe)], 
        cwd="../src/features/"
    )

    # 2. step make spectrogram images from cut recordings
    print('2. Making spectrogram images from cut recordings...')
    subprocess.call(
        ['python', 'make_spectrogram.py'], 
        cwd="../src/features/"
    )

    # 3. make filter (if it does not exist)
    print('3. Making filter...')
    subprocess.call(
        ['python', 'make_filter.py'], 
        cwd="../src/utils/"
    )

    # 4. apply filter to the images
    print('4. Applying filter to the images...')
    subprocess.call(
        ['python', 'apply_filter.py'], 
        cwd="../src/features/"
    )

def create_alternate_data(timeframe):
    # 1. step cut recordings
    print('1. Cutting recordings')
    subprocess.call(
        ['python', 'cut_original_recordings.py', str(timeframe), 'alternative'], 
        cwd="../src/features/"
    )

    # 2. step make spectrogram images from cut recordings
    print('2. Making spectrogram images from cut recordings...')
    subprocess.call(
        ['python', 'make_spectrogram.py', 'alternative'], 
        cwd="../src/features/"
    )

    # 4. apply filter to the images
    print('4. Applying filter to the images...')
    subprocess.call(
        ['python', 'apply_filter.py', 'alternative'], 
        cwd="../src/features/"
    )
```

```python
for timeframe in timeframes:
    print('Timeframe is %s ms' % (str(timeframe)))
    
    create_data(timeframe)
    
    subprocess.call(
        ['python', 'sort_train_validation_test_data.py', 'other'], 
        cwd="../src/features/"
    )
    
    create_alternate_data(timeframe)
    
    # 7. predict model with other
    model_name_tail = str(timeframe) + 'ms_other'
    model_name = 'model_' + model_name_tail
    
    print('7. Predict dataset with other')
    subprocess.call(
        ['python', 'predict_model.py', model_name, str(timeframe) + 'ms_25fil_other', 'other'], 
        cwd="../src/models/"
    )

## 20. predict model with other
#model_name_tail = str(timeframe) + 'ms'
#model_name = 'model_' + model_name_tail
#print('10. Predict dataset')
#subprocess.call(
#    ['python', 'predict_model.py', model_name, str(timeframe) + 'ms_25fil'], 
#    cwd="../src/models/"
#)
```

```python
for timeframe in timeframes:
    print('Timeframe is %s ms' % (str(timeframe)))
    
    create_data(timeframe)
    
    # ------------- CREATE alternate data (real data)
    
    create_alternate_data()

    
    # ------------- with other
    
    # 5. distribute images to training and test set in processed data folder wtih other class
    print('5. Creating training, validation and test set...')
    subprocess.call(
        ['python', 'sort_train_validation_test_data.py', 'other'], 
        cwd="../src/features/"
    )
    
    # 6. train model with other
    print('6. Train dataset with other')
    model_name_tail = str(timeframe) + 'ms_other'
    subprocess.call(
        ['python', 'train_model.py', '8', model_name_tail], 
        cwd="../src/models/"
    )
    
    # 7. predict model with other
    model_name = 'model_' + model_name_tail
    print('7. Predict dataset with other')
    subprocess.call(
        ['python', 'predict_model.py', model_name, str(timeframe) + 'ms_25fil_other', 'other'], 
        cwd="../src/models/"
    )
    
    # -------------
    
    # 8. reset folders
    print('8. Creating training, validation and test set...')
    subprocess.call(
        ['python', 'sort_train_validation_test_data.py'], 
        cwd="../src/features/"
    )
              
    # --------------
    
    # 9. train model
    print('9. Train dataset')
    model_name_tail = str(timeframe) + 'ms'
    subprocess.call(
        ['python', 'train_model.py', '7', model_name_tail], 
        cwd="../src/models/"
    )
    
    # 20. predict model
    model_name = 'model_' + model_name_tail
    print('10. Predict dataset')
    subprocess.call(
        ['python', 'predict_model.py', model_name, str(timeframe) + 'ms_25fil'], 
        cwd="../src/models/"
    )
```
