```python
from sty import fg, bg, ef, rs, Rule


def get_colors(is_other=False):
    silence = lambda txt: ('%s%s %s %s%s') % (bg.white, fg.black, txt, fg.rs, bg.rs)
    other = lambda txt: ('%s%s %s %s%s') % (bg.black, fg.white, txt, fg.rs, bg.rs)
    tone_0 = lambda txt: ('%s%s %s %s%s') % (bg.blue, fg.black, txt, fg.rs, bg.rs)
    tone_1 = lambda txt: ('%s%s %s %s%s') % (bg.green, fg.black, txt, fg.rs, bg.rs)
    tone_2 = lambda txt: ('%s%s %s %s%s') % (bg.cyan, fg.black, txt, fg.rs, bg.rs)
    tone_3 = lambda txt: ('%s%s %s %s%s') % (bg.red, fg.black, txt, fg.rs, bg.rs)
    tone_4 = lambda txt: ('%s%s %s %s%s') % (bg.yellow, fg.black, txt, fg.rs, bg.rs)
    tone_5 = lambda txt: ('%s%s %s %s%s') % (bg.magenta, fg.black, txt, fg.rs, bg.rs)

    COLORS = [
        tone_0,
        tone_1,
        tone_2,
        tone_3,
        tone_4,
        tone_5,
        tone_0,
        tone_1,
        tone_2,
        tone_3,
        tone_4,
        tone_5,
        silence,
    ]
    
    COLORS_TXT = [
        ('tone 0 small', 'blue'), ('tone 1 small', 'green'), ('tone 2 small', 'cyan'), 
        ('tone 3 small', 'red'), ('tone 4 small', 'yellow'), ('tone 5 small', 'magenta'),
        ('tone 0 great', 'blue'), ('tone 1 great', 'green'), ('tone 2 great', 'cyan'), 
        ('tone 3 great', 'red'), ('tone 4 great', 'yellow'), ('tone 5 great', 'magenta'),
        ('silence', 'white'),
    ]

    return [list(COLORS), list(COLORS_TXT)]
    
def plot_color_legend(is_other=False):
    # color legend
    colors, colors_txt = get_colors(is_other)
    for i, color in enumerate(colors):
        print(colors_txt[i][0] + '=>' + color(colors_txt[i][1]))
    print('\n---------------\n\n') 
```

```python
def convert_label_to_index(convert_list):
    converted_list = []
    for val in convert_list:
        if val == 'm0':
            index = 0
        elif val == 'm1':
            index = 1
        elif val == 'm2':
            index = 2
        elif val == 'm3':
            index = 3
        elif val == 'm4':
            index = 4
        elif val == 'm5':
            index = 5
        elif val == 'v0':
            index = 6
        elif val == 'v1':
            index = 7
        elif val == 'v2':
            index = 8
        elif val == 'v3':
            index = 9
        elif val == 'v4':
            index = 10
        elif val == 'v5':
            index = 11
        else:
            index = 12
        converted_list.append(index)
    return converted_list
        

def format_output(predictions_list, is_other=False):
    formated_string = ''
    
    for prediction in predictions_list:    
        formated_string += get_colors(is_other)[0][prediction](str(prediction))
    
    return formated_string

def get_filename(timeframe):
    return 'model_' + str(timeframe) + 'ms_' + CLASS_NUM + '_classes.txt'
    
def get_file_path(timeframe):
    '''
    timeframe -> string of defined time frame (10, 15, 20, 25, 30, 40, 50)
    returns alternative data files
    '''
    return os.path.join(REAL_DATA_FILES_DIR, get_filename(timeframe))
```

```python
from joblib import load
import h5py

# data for m, data for m with other, data for v, data for v with other
file_path = get_file_path(timeframe)

rnd_clf = load(os.path.join(MODEL_DIR, 'silece_lower_random_forest_default_name2.joblib'))
# model name
print(bg.white + get_filename(timeframe) + bg.rs)

plot_color_legend()

# recordings
for filename in os.listdir(REAL_DATA_AMP):
    # name of a recording
    print('\t' + str(filename))
    file = h5py.File(os.path.join(REAL_DATA_AMP, filename), 'r')
    y_predicted = rnd_clf.predict(file['amplitudes'].value)
    print(format_output(convert_label_to_index(y_predicted)))
    print('\n')
    file.close()
```
